### Project Summary: Customer Satisfaction (CSAT) Prediction for E-commerce Customer Support

This project focuses on predicting **Customer Satisfaction Score (CSAT)** using machine learning techniques based on customer support interaction data from an e-commerce platform. By analyzing historical customer support data and building predictive models, organizations can identify factors affecting customer satisfaction and take proactive steps to improve service quality.

The dataset used in this project contains information related to customer support interactions such as support channel, issue category, product category, customer city, item price, connected handling time, agent shift, tenure bucket of agents, and customer remarks. The target variable for this project is the **CSAT Score**, which reflects the level of customer satisfaction after a support interaction. Predicting this score can help businesses understand the drivers of customer satisfaction and improve their support services.

The project began with **data understanding and exploration**. The dataset was examined to understand its structure, the types of variables present, and potential data quality issues such as missing values or duplicate records. Exploratory Data Analysis (EDA) was conducted using visualizations including bar charts, histograms, scatter plots, and box plots. For example, certain support channels, product categories, and handling times appeared to influence customer satisfaction levels. These insights helped guide further data preparation and modeling.

The next step involved **data preprocessing and feature engineering** to prepare the dataset for machine learning algorithms. This included handling missing values, removing irrelevant columns, encoding categorical variables into numerical format using label encoding, and scaling numerical variables where necessary. Proper preprocessing ensured that the dataset was clean and suitable for model training.

The project also included **textual data preprocessing** for the customer remarks column. Since textual data cannot be directly used by most machine learning algorithms, several natural language processing (NLP) steps were applied. These steps included converting text to lowercase, removing punctuation, eliminating stopwords, tokenizing the text, and performing lemmatization to reduce words to their root forms. After cleaning the text data, **TF-IDF (Term Frequency–Inverse Document Frequency)** vectorization was used to transform the textual information into numerical features that could be incorporated into the machine learning models.

In addition to exploratory analysis, **hypothesis testing** was conducted to statistically evaluate relationships between variables observed during EDA. Hypotheses related to factors such as support channel and agent shift influencing CSAT scores were tested using appropriate statistical methods. This helped validate whether the observed relationships were statistically significant.

For the machine learning phase, the dataset was split into **training and testing sets** to ensure proper model evaluation. Three regression models were implemented: **Linear Regression, Random Forest Regressor, and Gradient Boosting Regressor**. These models were trained on the prepared dataset and evaluated using performance metrics such as **Mean Absolute Error (MAE), Root Mean Squared Error (RMSE), and R² Score**. These evaluation metrics measure prediction accuracy and the model's ability to explain variability in the target variable.

Among the models tested, the **Random Forest Regressor demonstrated the best performance**, achieving lower prediction errors and a higher R² score compared to the other models. Random Forest is an ensemble learning algorithm that builds multiple decision trees and combines their predictions to improve accuracy and reduce overfitting. Feature importance analysis was conducted to understand which variables contributed most to predicting customer satisfaction. Factors such as handling time, support channel, product category, and agent-related attributes were identified as important predictors of CSAT scores.

In conclusion, this project demonstrates how machine learning techniques can be used to analyze customer support data and predict customer satisfaction levels. The results highlight key factors that influence customer satisfaction and provide valuable insights for improving customer support operations. By understanding these factors, businesses can optimize their support processes, reduce resolution time, and enhance overall customer experience.

### Hypothetical Statement - 1

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

Null Hypothesis (H0):
There is no significant difference in CSAT scores between different support channels.

Alternative Hypothesis (H1):
There is a significant difference in CSAT scores between different support channels.

#### 2. Perform an appropriate statistical test.

````python
# Perform Statistical Test to obtain P-Value

from scipy.stats import ttest_ind

inbound=df[df['channel_name']=='Inbound']['CSAT Score']

outbound=df[df['channel_name']=='Outcall']['CSAT Score']

ttest_ind(inbound,outbound)
````

##### Which statistical test have you done to obtain P-Value?

Independent T-Test

##### Why did you choose the specific statistical test?

The independent T-test is used to compare the mean CSAT scores between two independent groups (Inbound and Outbound channels). It helps determine whether the difference between the two groups is statistically significant.

### 1. Handling Missing Values

````python
# Handling Missing Values & Missing Value Imputation

# Check missing values
df.isnull().sum()

# For numerical columns fill with median
df['Item_price'] = df['Item_price'].fillna(df['Item_price'].median())
df['connected_handling_time'] = df['connected_handling_time'].fillna(df['connected_handling_time'].median())

# For categorical columns fill with mode
df['Product_category'] = df['Product_category'].fillna(df['Product_category'].mode()[0])
df['Customer_City'] = df['Customer_City'].fillna(df['Customer_City'].mode()[0])

# Verify again
df.isnull().sum()
````

#### What all missing value imputation techniques have you used and why did you use those techniques?

Missing values were handled using simple imputation techniques.

For numerical variables such as Item_price and connected_handling_time, missing values were replaced using the median value. The median is preferred because it is less affected by extreme values and provides a stable estimate.

For categorical variables such as Product_category and Customer_City, missing values were replaced using the mode (most frequent value). This helps maintain the distribution of the categorical variables.

Handling missing values ensures that the dataset becomes complete and ready for machine learning algorithms.

### 4. Textual Data Preprocessing
(It's mandatory for textual dataset i.e., NLP, Sentiment Analysis, Text Clustering etc.)

#### 1. Expand Contraction

````python
# Expand Contraction

# Expand Contraction

import contractions

df['Customer Remarks'] = df['Customer Remarks'].astype(str).apply(contractions.fix)

df['Customer Remarks'].head()
````

#### 2. Lower Casing

````python
# Lower Casing

df['Customer Remarks'] = df['Customer Remarks'].str.lower()

df['Customer Remarks'].head()
````

#### 3. Removing Punctuations

````python
# Remove Punctuations

import string

df['Customer Remarks'] = df['Customer Remarks'].str.replace(f"[{string.punctuation}]", "", regex=True)

df['Customer Remarks'].head()
````

#### 4. Removing URLs & Removing words and digits contain digits.

````python
# Remove URLs & Remove words and digits contain digits
import re

df['Customer Remarks']=df['Customer Remarks'].apply(lambda x:re.sub(r'http\S+', '', x))

df['Customer Remarks'] = df['Customer Remarks'].apply(lambda x: re.sub(r'\w*\d\w*', '', x))

df['Customer Remarks'].head()
````

#### 5. Removing Stopwords & Removing White spaces

````python
# Remove Stopwords

import nltk

from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words=set(stopwords.words('english'))

df['Customer Remarks'] = df['Customer Remarks'].apply(
    lambda x: " ".join([word for word in x.split() if word not in stop_words])
)

df['Customer Remarks'].head()
````

````python
# Remove White spaces

df['Customer Remarks']=df['Customer Remarks'].str.strip()

df['Customer Remarks'].head()
````

#### 6. Rephrase Text

````python
# Rephrase Text

df['Customer Remarks'] = df['Customer Remarks'].apply(lambda x: " ".join(x.split()))

df['Customer Remarks'].head()
````

#### 7. Tokenization

````python
# Tokenization

from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

df['tokens'] = df['Customer Remarks'].apply(word_tokenize)

df['tokens'].head()
````

#### 8. Text Normalization

````python
# Normalizing Text (i.e., Stemming, Lemmatization etc.)

from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer = WordNetLemmatizer()

df['tokens'] = df['tokens'].apply(lambda words: [lemmatizer.lemmatize(word) for word in words])

df['tokens'].head()
````

##### Which text normalization technique have you used and why?

Lemmatization was used as the text normalization technique.

Lemmatization converts words to their base or dictionary form, which helps reduce variations of the same word. For example, words like "running", "runs", and "ran" are converted to "run".

#### 10. Text Vectorization

````python
# 10. Vectorizing Text (TF-IDF)

from sklearn.feature_extraction.text import TfidfVectorizer

# Ensure the column is string
df['Customer Remarks'] = df['Customer Remarks'].astype(str)

# Replace empty strings with placeholder
df['Customer Remarks'] = df['Customer Remarks'].replace('', 'no_text')

# Remove rows that still have empty text
df['Customer Remarks'] = df['Customer Remarks'].fillna('no_text')

# Initialize TF-IDF
tfidf = TfidfVectorizer(max_features=500)

# Fit transform
X_text = tfidf.fit_transform(df['Customer Remarks'])

# Convert to array
X_text = X_text.toarray()

# Convert to dataframe
tfidf_df = pd.DataFrame(X_text, columns=tfidf.get_feature_names_out())

# Display first rows
tfidf_df.head()
````

##### Which text vectorization technique have you used and why?

The TF-IDF (Term Frequency – Inverse Document Frequency) vectorization technique was used to convert textual data into numerical form.

TF-IDF measures the importance of a word in a document relative to the entire dataset. Words that appear frequently in a particular document but rarely across the dataset receive higher weights, while commonly occurring words receive lower weights.

This technique is suitable for this project because it helps capture meaningful keywords from customer remarks while reducing the impact of very common words. By converting the text into numerical vectors, machine learning models can process and learn patterns from the textual customer feedback.
